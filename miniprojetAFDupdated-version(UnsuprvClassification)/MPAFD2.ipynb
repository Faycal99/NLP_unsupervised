```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df =pd.read_csv("store_data.csv",header=None)
#pandas considerer la premiere ligne comme un header
```

```python
df
```

```python
#Nettoyage et praparation des donnes textuelles
records=[]
```

```python
df.shape
```

```python
for i in range(7501):
    records.append([str(df.values[i,j]) for j in range(20) if str(df.values[i,j]) != "nan" ])#la plus long transaction contient 20 articles
    #on supprime les nan
```

```python
records[1]
```

```python
len(records)
```

```python
#Application de l algorithme apriyori
Association_rules=apriori(records,min_support=0.0045,min_confidence=0.1,min_lift=3,min_length=2)
```

```python
print(type(Association_rules))
```

```python
#casting de ce generatot vers une list
Association_rules_list=list(Association_rules)
```

```python
list(Association_rules)
```

```python
 for item in Association_rules_list:
        print(item,"\n\n\n")
```

```python
for item in Association_rules_list:
    pair=item [0]
    items=[x for x in pair]
    print("Rule:" +items[0]+"-->"+items[1])
    print("Support:"+str(item[1]))
    print("Confidence:"+str(item[2][0][2]))
    print("Lift:"+str(item[2][0][2]))
    print("==========================")
```
